# 2. Sử lý dữ liệu

```python
df1["released"]= df1["released"].str.split("(", n = 1, expand = True)
```

```python
df1['released'] = pd.to_datetime(df1['released'])
print(df1['released'].dtypes)
```

```python
# Chuyển thành công cột "released" thành kiểu dữ liệu datetime
```

# 3.2Tổng doanh thu của phim theo năm

```python
df1.groupby('year').mean()['revenue']
```

# 3.3 Mối quan hệ giữa số lượng phim phát hành theo các tháng trong 1 năm

```python
df1['Month_released']=df1['released'].dt.month
df1
```

```python
df1.groupby('Month_released').count()['name']
```

```python
month_release = df1.groupby('Month_released').count()['name']
month = [month for month, v in month_release.items()]
plt.bar(month,month_release)
plt.show()
```

```python
#  Không có quá nhiều sự chênh lệch về số lượng phim phát hành giữa các tháng trong năm.
#  điều đó cho thấy, yếu tố thời gian không quá ảnh hưởng đến số lượng phim phát hành
```
